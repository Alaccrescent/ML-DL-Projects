Model building:

Let us build the model using the squareMeters column first.

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
x=data.drop(["price"],axis=1).copy()
y=data["price"].copy()

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=100)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
y_tr_pred=model.predict(x_train)
print("Train R2 score",r2_score(y_train,y_tr_pred))

y_pred=model.predict(x_test)
print("*"*75)
print("Test R2 score:",r2_score(y_test,y_pred))
```

```python
errors=y_train-y_tr_pred
sns.scatterplot(x=y_tr_pred,y=errors)
plt.xlabel("Fitted values")
plt.ylabel("Residuals")
plt.title("Residuals vs y_fitted values")
plt.show()
```

```python
import statsmodels.api as sm
import scipy.stats as stats
import warnings
warnings.simplefilter("ignore")
sm.qqplot(errors, line="q")
plt.show()
```

Observations:

As observed from the plot the residuals are homoscedastic in nature. Also the residuals are normally distributed.
